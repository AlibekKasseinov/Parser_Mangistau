```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
file_path = '/Users/alishershaimuran/Desktop/итог 20.02 вместе с кол.xlsx'
df = pd.read_excel(file_path)
id_list = df['business_id'].astype(str).str.zfill(12)
df['id_list']=id_list
```

```python
df
```

```python

data_akb = []

for id in id_list:
    url = f'https://pk.uchet.kz/search/bin/{id}'

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses
        soup = BeautifulSoup(response.text, 'html.parser')

        address_item_elements = soup.find_all('div', class_='info-block__object-block')

        if address_item_elements:
            # Access the first element (index 0)
            city_div = address_item_elements[0]

            # Now you can manipulate the elements as needed
            city_text = city_div.get_text()
            print(city_text)
            data_akb.append(city_text)
        else:
            city_text = "НЕТ ССЫЛКИ"
            print(city_text)
            data_akb.append(city_text)

    except requests.exceptions.RequestException as e:
        error_message = f"Error fetching data for id {id}: {e}"
        print(error_message)
        data_akb.append(error_message)

# Continue with the rest of your code
```

```python
new_addresses = []

words_to_remove = ["УЛИЦА", "УЛ ", "улица ", "Жилой массив"]

for address in data_akb:
    if "С.О." in address:
        for word in words_to_remove:
            index = address.find(word)
            if index != -1:
                new_address = address[:index].strip()  
                new_addresses.append(new_address)
                break  
    else:
        new_addresses.append(address)  

# Выводим результаты для проверки
for new_address in new_addresses:
    print(new_address)
```
